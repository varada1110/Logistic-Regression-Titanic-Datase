```python
import numpy as np
import pandas as pd
data = pd.read_csv("training_titanicLR.csv",delimiter=",")
data.info()
data
dataset = pd.read_csv("test_titanicLR.csv",delimiter=",")
dataset.info()
dataset
```

```python
data.drop(['Sex','Embarked','Name','Ticket','Cabin'],axis=1,inplace=True)
data.head()
```

```python
dataset.drop(['Sex','Embarked','Name','Ticket','Cabin'],axis=1,inplace=True)
dataset.head()
```

```python
data.isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
data.Age.fillna(data.Age.mean(),inplace=True)
dataset.Age.fillna(data.Age.mean(),inplace=True)
```

```python
data.head()
```

```python
dataset.head()
```

```python
x = data.drop("Survived",axis=1)
y = data["Survived"]
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(x,y,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
y_pred = clf.predict(dataset)
```

```python
y_pred
np.savetxt('titanic_ypred.csv',y_pred,delimiter=",")
```

```python
clf.predict_proba(dataset)
```
